```python
import numpy as np
np.set_printoptions(precision=4, suppress=True)
```

```python
x1=1
x2=0
w1=-0.3
w2=0.6
bias=0.2
eta=0.1 #step-size
d=0.8 #desired output

def activation(A):
    return 1/(1.0+np.exp(-A))

for i in range(0, 3000):
    A=w1*x1 + w2*x2 + bias
    y=activation(A)
    delta = y*(1-y)*(d-y)
    w1=w1 + eta*delta*x1
    w2=w2 + eta*delta*x2
    bias=bias + eta*delta*1
    print("w1, w2, bias, delta")
    print(w1, w2, bias, delta)
```
